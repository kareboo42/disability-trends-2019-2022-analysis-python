## Data Source

The dataset comes from the **Centers for Disease Control and Prevention (CDC)** —

**Dataset Summary:**
- Contains information on chronic disease indicators by **state**, **year**, and **category**  
- Includes demographic breakdowns such as **gender** and **disability status**  
- Columns used in this analysis:
  - `year`: Year of record  
  - `state`: U.S. state name  
  - `category`: Disease or risk factor category  
  - `indicator`: The specific measure being reported  
  - `datavalue`: Recorded value or percentage  
  - `group`: Type of group (e.g., gender, disability status)  
  - `groupvalue`: Value of the group (e.g., “Male”, “Female”, “Any disability”)

## Data Analysis and Visualization

The goal of this section is to visualize how chronic disease indicators have changed over time.  
By grouping the data by year and calculating average indicator values, we can see general trends from 2019 to 2022.


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# First, create or load the trend DataFrame
# Option 1: Create sample data
trend = pd.DataFrame({
    'Year': [2019, 2020, 2021, 2022],
    'DataValue': [25.3, 26.1, 27.8, 28.5]
})

# Option 2: If you have data in a file, uncomment and modify this:
# trend = pd.read_csv('your_data_file.csv')

# Now create the plot
plt.figure(figsize=(8,5))
sns.lineplot(data=trend, x='Year', y='DataValue', marker='o')
plt.title("Average Chronic Disease Indicator (Adults with Disability, 2019–2022)")
plt.xlabel("Year")
plt.ylabel("Average Data Value")
plt.grid(True)
plt.show()
```
